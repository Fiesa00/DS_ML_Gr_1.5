### Merge

#### ACHTUNG! Hier werden die Dataframes gemerged, Aber das data_merged_new wird nicht berücksichtigt!

```python


# Sicherstellen, dass "Datum" das gleiche Format hat
for df in [df_inflation, df_heimspiele, df_maerkte, df_passagiere, df_temperatur, df_umsatz,df_frei, df_wetter, df_kiwo]:
    df['Datum'] = pd.to_datetime(df['Datum'])

# Filter: Daten im Zeitraum 2013-07-01 bis 2019-07-31.
start_date = pd.Timestamp('2013-07-01')
end_date = pd.Timestamp('2019-07-31')
dfs = [df[(df['Datum'] >= start_date) & (df['Datum'] <= end_date)] for df in [df_inflation, df_heimspiele, df_maerkte, df_passagiere, df_temperatur, df_umsatz,df_frei,df_wetter, df_kiwo]]

# Alle DataFrames iterativ mergen
df_merged = dfs[0]

for df in dfs[1:]:
    df_merged = pd.merge(df_merged, df, on="Datum", how="outer")  # "outer" verbindet alle Daten

# Nach Datum sortieren
df_merged.sort_values('Datum', inplace=True)

# Liste der Spalten, die gelöscht werden sollen
zu_loeschende_spalten = ['Brot_W1', 'Broetchen_W2','Croissant_W3','Konditorei_W4','Kuchen_W5','Saisonbrot_W6']

# Spalten löschen
df_merged= df_merged.drop(columns=zu_loeschende_spalten, errors='ignore')

# Speichern als CSV
output_csv = "/workspaces/DS_ML_Gr_1.5/2_BaselineModel/merged_data.csv"
df_merged.to_csv(output_csv, index=False)

print(df_merged.head())

print(f"Die Datei wurde erfolgreich unter '{output_csv}' gespeichert.")
```

```python
def encode_weekdays(df):
    """
    Erstellt für jeden Wochentag eine neue Spalte mit 1 oder 0,
    basierend auf dem Datum in der 'Datum'-Spalte.
    
    Parameters:
    df (pandas.DataFrame): DataFrame mit einer 'Datum'-Spalte im Format 'YYYY-MM-DD'
    
    Returns:
    pandas.DataFrame: Original DataFrame mit zusätzlichen Wochentags-Spalten
    """
    # Stelle sicher, dass die Datum-Spalte als datetime-Typ vorliegt
    df['Datum'] = pd.to_datetime(df['Datum'])
    
    # Erstelle für jeden Wochentag eine neue Spalte
    weekdays = {
        'Monday': 0,
        'Tuesday': 1,
        'Wednesday': 2,
        'Thursday': 3,
        'Friday': 4,
        'Saturday': 5,
        'Sunday': 6
    }
    
    for day, day_num in weekdays.items():
        df[day] = (df['Datum'].dt.dayofweek == day_num).astype(int)
    
    return df
```

```python
df_all_codiert = encode_weekdays(df_merged_dropped)
```
